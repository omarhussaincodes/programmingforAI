```python
import math

# Parent class


class Shapes:
    def __init__(self, name, shape_type):
        self.name = name
        self.shape_type = shape_type
    # this method will be overridden in subclass or child classes

    def get_area(self):
        pass

# subclass 1


class Rectangle(Shapes):
    def __init__(self, name, shape_type, length, width):
        super().__init__(name, shape_type)
        self.length = length
        self.width = width

    def get_area(self):
        return (self.length * self.width)

    def get_perimetre_rectangle(self):
        return 2 * (self.length + self.width)

# subclass 2


class Triangle(Shapes):
    def __init__(self, name, shape_type, base, height):
        super().__init__(name, shape_type)
        self.base = base
        self.height = height

    def get_area(self):
        return (1/2 * (self.base * self.height))

# subclass 3


class Circle(Shapes):
    def __init__(self, name, shape_type, radius):
        super().__init__(name, shape_type)
        self.radius = radius

    def get_area(self):
        return (math.pi * math.pow(self.radius, 2))

    def get_circle_diameter(self):
        return 2*(self.radius)

    def get_circle_circumference(self):
        return (2 * math.pi * self.radius)

# subclass 4


class Sphere(Shapes):
    def __init__(self, name, shape_type, radius):
        super().__init__(name, shape_type)
        self.radius = radius

    def get_area(self):
        return (4 * math.pi * math.pow(self.radius, 2))

    def get_sphere_diameter(self):
        return 2*(self.radius)

    def get_sphere_circumference(self):
        return (2 * math.pi * self.radius)

    def get_sphere_volume(self):
        return 4/3 * (math.pi * math.pow(self.radius, 3))


# Create objects of each subclass and calculate their areas
rectangle = Rectangle("Rectangle", "2D", 4.25, 5.75)
triangle = Triangle("Triangle", "2D", 2.0, 4.0)
circle = Circle("Circle", "2D", 5.0)
sphere = Sphere("Sphere", "3D", 6.0)

# Calculating and printing the areas
print(f"Area of {rectangle.name} of type {rectangle.shape_type} is : {rectangle.get_area()}")
print(f"Area of {triangle.name} of type {triangle.shape_type} is : {triangle.get_area()}")
print(f"Area of {circle.name} of type {circle.shape_type} is : {circle.get_area()}")
print(f"Area of {sphere.name} of type {sphere.shape_type} is : {sphere.get_area()}")

print("============================================================================")

# calculate and print result of some subclass methods
print(f"Volume of {sphere.name} of type {sphere.shape_type} is : {sphere.get_sphere_volume()}")
print(f"Diametre of {circle.name} of type {circle.shape_type} is : {circle.get_circle_diameter()}")
print(f"Perimetre of {rectangle.name} of type {rectangle.shape_type} is : {rectangle.get_perimetre_rectangle()}")
```
